# Extração de Dados

## Listagem de todos os times que estão no campeonato

```python
def get_times(partidas:pd.DataFrame) -> list[str]:
  '''
  Retorna uma lista com todos os times que
  participaram do campeonato.
  '''
  times = list()
  # É lida apenas a primeira coluna pois nela
  # já estão todos os times
  for row in partidas[1]:
    times.append(row['clubs']['home'])
    times.append(row['clubs']['away'])

  times.sort()

  return times
```

```python
times = get_times(partidas)
```

## Listagem de Pontos, Vitórias, Saldo de gols e Trocas de Técnico por time

```python
def get_resultados(partidas:pd.DataFrame, times:list[str]) -> list[dict]:
  '''
  Retorna uma lista de dicionários, cada dicionário com os
  dados de um determinado time, contendo o nome do time, total de
  pontos, vitórias, saldo de gols e total de trocas de técnico.
  Os dicionários estão ordenados pela classificação do time.
  '''
  def get_resultado(partida:dict, pos_time:str, pos_advs:str) -> tuple[int]:
    '''
    Retorna o total de pontos ganhos e a
    diferença de gols de uma partida

    A 'pos_time' e 'pos_advs' indica qual time
    é de casa(home) e qual é visitante(away)
    '''
    gols_time = int( partida['goals'][pos_time] )
    gols_advs = int( partida['goals'][pos_advs] )

    saldo = gols_time - gols_advs

    if   gols_time > gols_advs: # vitória
      return (3, saldo)
    elif gols_time < gols_advs: # derrota
      return (0, saldo)
    else:                       # empate
      return (1, saldo)


  def get_primeiro_tecnico(partidas:pd.DataFrame, time:str) ->str:
    '''
    Retorna o nome do técnico de um determinado time
    durante a primeira rodada do campeonato
    '''
    for partida in partidas[1]:

      if   partida['clubs']['home'] == time:
        return partida['coach']['home']
      elif partida['clubs']['away'] == time:
        return partida['coach']['away']


  resultados = list()

  # Estes laços percorrem a lista de times e para cada time
  # percorre todas as rodadas do campeonato e em cada rodada
  # encontra a partida que o time jogou. Ao encontrar a partida
  # soma os pontos da partida ao total de pontos, soma a diferença
  # de gols ao saldo de gols, se o time ganhou, incrementa 1 a
  # variável vitorias, e se houve troca de técnico, incrementa em 1
  # a variável trocas_tecnicos.

  for time in times:

    total_pontos   = 0
    saldo_gols     = 0
    vitorias       = 0
    trocas_tecnico = 0
    tecnico = get_primeiro_tecnico(partidas, time)

    for rodada in partidas:
      for partida in partidas[rodada]:

        jogou = False

        if   partida['clubs']['home'] == time:
          pos_time = 'home'
          pos_advs = 'away'
          jogou = True

        elif partida['clubs']['away'] == time:
          pos_time = 'away'
          pos_advs = 'home'
          jogou = True

        if jogou:
          pontos, saldo = get_resultado(partida, pos_time, pos_advs)

          total_pontos += pontos
          saldo_gols += saldo
          if pontos == 3:
            vitorias += 1

          novo_tecnico = partida['coach'][pos_time]
          if tecnico != novo_tecnico:
            trocas_tecnico += 1
            tecnico = novo_tecnico

          # Como cada time joga apenas uma vez por rodada,
          # após achar um jogo com certo time, já se pode
          # partir para a próxima rodada
          break

    resultados.append({
        'Time': time,
        'Pontos': total_pontos,
        'Vitórias': vitorias,
        'Saldo_gols': saldo_gols,
        'Trocas_Técnicos': trocas_tecnico
    })


  # Ordena os resultados pelo total de pontos e com
  # número de vitórias como criério de desempate


  if PANDAS: # Com Pandas

    resultados = pd.DataFrame(resultados)

    for i in range(resultados.shape[0]):
      for j in range(i, resultados.shape[0]):

        if resultados['Pontos'].iloc[i] < resultados['Pontos'].iloc[j]:

          temp = resultados.iloc[i]
          resultados.iloc[i] = resultados.iloc[j]
          resultados.iloc[j] = temp

        elif resultados['Pontos'].iloc[i] == resultados['Pontos'].iloc[j]:
          if resultados['Vitórias'].iloc[i] < resultados['Vitórias'].iloc[j]:

            temp = resultados.iloc[i]
            resultados.iloc[i] = resultados.iloc[j]
            resultados.iloc[j] = temp

  else: # Sem Pandas

    for i in range(len(resultados)):
      for j in range(i, len(resultados)):

        if resultados[i]['Pontos'] < resultados[j]['Pontos']:

          temp = resultados[i]
          resultados[i] = resultados[j]
          resultados[j] = temp

        elif resultados[i]['Pontos'] == resultados[j]['Pontos']:
          if resultados[i]['Vitórias'] < resultados[j]['Vitórias']:

            temp = resultados[i]
            resultados[i] = resultados[j]
            resultados[j] = temp


  return resultados
```

## Listagem das Maiores Durações de Técnico, Trocas de Técnicos e os resultados após as Trocas

```python
def get_resultados_apos_troca(partidas:pd.DataFrame, times:list[str]) -> tuple[ list[dict], list[ list[str] ], list[int] ]:
  '''
  Retorna uma tupla com 3 elementos:
  1.Lista com a maior duração de técnico de cada time.
  2.Lista com todas as trocas de técnico ocorridas durante
  o campeonato.
  3.Lista com o total de vitórias, derrotas e empates
  ocorridos na partida seguinte a uma troca de técnico.
  '''

  vitorias = 0
  derrotas = 0
  empates  = 0
  trocas = list()
  max_duracoes = list()

  for time in times:
    # Variáveis referente as durações dos técnicos
    duracao  = 1
    duracoes = list()
    # Encontra o primeiro técnico
    for partida in partidas[1]:

      if   partida['clubs']['home'] == time:
        tecnico = partida['coach']['home']
        break

      elif partida['clubs']['away'] == time:
        tecnico = partida['coach']['away']
        break

    # Analisa as rodadas seguintes
    for rodada in partidas.columns[1:]:
      for partida in partidas[rodada]:

        jogou = False

        if   partida['clubs']['home'] == time:
          pos_time = 'home'
          pos_advs = 'away'
          jogou = True

        elif partida['clubs']['away'] == time:
          pos_time = 'away'
          pos_advs = 'home'
          jogou = True

        if jogou:

          novo_tecnico = partida['coach'][pos_time]

          if tecnico != novo_tecnico:
            # Guarda o registro da duração
            duracoes.append(duracao)
            duracao = 1
            # Cria um novo registro de troca
            trocas.append([time, tecnico, novo_tecnico])
            # Atualiza o nome do técnico
            tecnico = novo_tecnico
            gols_time = partida['goals'][pos_time]
            gols_advs = partida['goals'][pos_advs]
            # Analisa o resultado da partida
            if   gols_time > gols_advs:
              vitorias += 1
            elif gols_time < gols_advs:
              derrotas += 1
            else:
              empates  += 1
          else:
            duracao += 1

          break

    duracoes.append(duracao)
    max_duracoes.append(max(duracoes))


  if PANDAS: # Com Pandas

    duracoes = pd.DataFrame({'Time':times, 'Max_Duração':max_duracoes})
    trocas = pd.DataFrame(trocas, columns=['Time', 'Técnico_ant', 'Técnico_novo'])

  else: # Sem Pandas
    duracoes = [{'Time':time, 'Max_Duração':duracao} for time, duracao in zip(times, max_duracoes)]


  contagem = [ vitorias ,  derrotas ,  empates ]

  return (duracoes, trocas, contagem)
```

```python
duracoes, trocas, contagem = get_resultados_apos_troca(partidas, times)
```

## Ajustes finais

```python
# Junção das tabelas

if PANDAS: # Com Pandas

  resultados = pd.merge(resultados, duracoes[['Time', 'Max_Duração']], on='Time', how='inner')

else: # Sem Pandas

  for duracao in duracoes:
    for resultado in resultados:
      if duracao['Time'] == resultado['Time']:
        resultado['Max_Duração'] = duracao['Max_Duração']
        break
```

```python
if PANDAS:
  # Adiciona 1 em cada índice para representar a classificação
  resultados.index = [i+1 for i in resultados.index]
```
