#__Exercício 3:__ Análise de Dados de Compras Online


__Dataset:__ Disponível no Kaggle como _Brazilian E-Commerce Public Dataset by Olist._

__Questões:__

```python
from datetime import datetime
```

```python
df = pd.read_csv('olist_orders_dataset.csv')
df
```

#### __1-__ Determine quantos pedidos foram feitos em cada ano.

```python
# Criando uma coluna para o ano
df['year'] = df['order_purchase_timestamp'].apply(lambda x: x[:4])
df[['order_purchase_timestamp', 'year']]
```

```python
# Exibindo a contagem de vendas por ano
df['year'].value_counts()
```

#### __2-__ Calcule o tempo médio de entrega (diferença entre order_purchase_timestamp e order_delivered_customer_date).

```python
df[['order_purchase_timestamp', 'order_delivered_customer_date']]
```

```python
df = df.dropna()
```

```python
def converte_para_datetime(data):

  ano = int(data[:4])
  mes = int(data[5:7])
  dia = int(data[8:10])
  hor = int(data[11:13])
  min = int(data[14:16])
  seg = int(data[17:])

  return datetime(ano, mes, dia, hor, min, seg)
```

```python
df['order_purchase_timestamp'] = df['order_purchase_timestamp'].apply(converte_para_datetime)
```

```python
df['order_delivered_customer_date'] = df['order_delivered_customer_date'].apply(converte_para_datetime)
```

```python
# Média do tempo de entrega
df['order_delivered_customer_date'].mean() - df['order_purchase_timestamp'].mean()
```

#__Exercício 4:__ Análise de Dados de Filmes do IMDb



    
__Dataset:__ Disponível no Kaggle como _IMDb movies extensive dataset_.

```python
from datetime import date
```

```python
movies = pd.read_csv('IMDb movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('IMDb ratings.csv')
ratings.head()
```

#### __1-__ Exiba os títulos e anos dos 10 filmes mais recentes no dataset.

```python
def converte_para_date(data):

  try:
    ano = int(data[:4])
  except ValueError:
    ano = 1900
  try:
    mes = int(data[5:7])
  except ValueError:
    mes = 1
  try:
    dia = int(data[8:10])
  except ValueError:
    dia = 1

  return date(ano, mes, dia)
```

```python
movies['date_published'] = movies['date_published'].apply(converte_para_date)
```

```python
movies_10_new = movies.sort_values('date_published', ascending=False)[:10]
movies_10_new[['title', 'year']]
```

#### __2-__ Calcule a duração média dos filmes.

```python
round(movies['duration'].mean(), 2)
```

#### __3-__ Identifique o filme com a maior quantidade de votos (votes).

```python
movies[movies['votes']==movies['votes'].max()]
```

#### __4-__ Determine quantos filmes possuem uma classificação (avg_vote) superior a 8.

```python
len(movies[movies['avg_vote'] > 8])
```
